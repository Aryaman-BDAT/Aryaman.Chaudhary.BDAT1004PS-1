<h2>Question 3</h2>
<h5>Heron's Formula and area of a Triangle</h5>

```
using System;

namespace HeronsFormula
{
    class Program
    {
        static void Main(string[] args)
        {
            //declaring variables to store values for the triangle
            double s, a, b, c, areaTriangle;

            //Getting input of all the 3 sides of the triangle
            Console.WriteLine("Enter the values of a, b, c \n");
            a = Convert.ToDouble(Console.ReadLine());
            b = Convert.ToDouble(Console.ReadLine());
            c = Convert.ToDouble(Console.ReadLine());

            //Calculating value of s
            s = (a + b + c) / 2;

            //Calculating area of triangle
            //math.sqrt is a funtion to get square root os any value
            areaTriangle = Math.Sqrt(s * (s - a) * (s - b) * (s - c));

            //printing area of triangle
            Console.WriteLine("Area of triangle is " + areaTriangle);
            Console.ReadLine();
        }
    }
}
```

<h2>Question 4</h2>
<h5>Odd Even Sort Array</h5>

```
using System;

namespace OddEvenSortArray
{
    class Program
    {
        static void Main(string[] args)
        {
            // Declatarion of 3 arrays to store values of unsorted array, even numbers array and odd numbers array respectively
            int[] collector = new int[10];
            int[] evenArray = new int[10];
            int[] oddArray = new int[10];
            // variables to be used in running loops
            int a, b = 0, c = 0, i;


            //Getting an idea of how many numbers are to be stored in the array to be sorted later
            Console.Write("Input the number of elements to be stored in the array :");
            i = Convert.ToInt32(Console.ReadLine());

            // Storing values one by one in the collector array
            Console.Write("Input {0} elements in the array :\n", i);
            for (a = 0; a < i; a++)
            {
                Console.Write("element - {0} : ", a);
                collector[a] = Convert.ToInt32(Console.ReadLine());
            }

            //Sorting elements of collector array into odd and even and then storing them in the respective arrays
            for (a = 0; a < i; a++)
            {
                if (collector[a] % 2 == 0)
                {
                    evenArray[b] = collector[a];
                    b++;
                }
                else
                {
                    oddArray[c] = collector[a];
                    c++;
                }
            }

            //Printing the even and odd elements
            Console.Write("\nThe Even elements are : \n");
            for (a = 0; a < b; a++)
            {
                Console.Write("{0} ", evenArray[a]);
            }

            Console.Write("\nThe Odd elements are :\n");
            for (a = 0; a < c; a++)
            {
                Console.Write("{0} ", oddArray[a]);
            }
            Console.ReadLine();
        }
    
    }
}
```

<h2>Question 5</h2>
<h5>Point inside a Rectacngle</h5>

```
using System;

namespace PointInsideRectangle
{
    class Program
    {
        static void Main(string[] args)
        {
            //declaration of variables to be considered as the points
            Double x, x1, x2, y, y1, y2;

            //inout for variables
            Console.WriteLine("Input the point you want to find and the lower left and upper right corner of the rectangle \n");
            x = Convert.ToDouble(Console.ReadLine());
            y = Convert.ToDouble(Console.ReadLine());
            x1 = Convert.ToDouble(Console.ReadLine());
            y1 = Convert.ToDouble(Console.ReadLine());
            x2 = Convert.ToDouble(Console.ReadLine());
            y2 = Convert.ToDouble(Console.ReadLine());

            //function call for intake of values to be calculates and getting value in form of bool
            bool decision = Inside(x, y, x1, y1, y2, y2);

            //Printing bool decision
            Console.WriteLine(decision);
            Console.ReadLine();
        }


        // function def 
        static bool Inside(Double a, Double b, Double a1, Double b1, Double a2, Double b2)
        {
            if (a > a1 && a < a2 && b > b1 && b < b2)
                return true;
            else
                return false;
        }
    }
}
```
